```python
'''
寻找水仙花数
'''
for num in range(100,1000):
    h = num // 100
    unit = num % 10
    ten = num // 10 % 10
    if num == h ** 3 + ten ** 3 + unit ** 3 :
        print("%d" %num)
    
```

```python
'''
寻找完美数
1-9999
小于本身的因数
'''

for num in range(1, 10000):
    sum = 0
    for factor in range(1, num):
        if num % factor == 0:
            sum += factor
            #if factor > 1 and num / factor != factor:
             #   sum += num / factor 不明白这是什么意思
    if num == sum:
        print(num)
```

```python
'''
百鸡百钱

一只公鸡cock五个钱, 一只母鸡hen2个钱, 2个小鸡chicks一个钱, 100个钱买100个鸡
问: 公鸡, 母鸡, 小鸡 各多少个
'''
for cock in range(20):
    for hen in range(50):
        chicks = 100 - cock - hen
        if 100 == cock * 5 + hen * 2 + hen // 2 :
            print('%d cock %d hen %d chicks' %(cock, hen, chicks))
            x = cock * 5 + hen * 2 + hen / 2 
            print(x)
```

```python
'''
craps赌博问题 

玩家摇骰子,如果第一次 摇出7点或者11点, 玩家胜
如果摇出2点, 3点, 12点, 庄家胜, 其他情况游戏继续
玩家再次摇色子, 如果摇出7点, 庄家胜
如果摇出第一次摇的点数, 玩家胜
否则游戏继续, 玩家继续摇骰子
玩家进入游戏时有1000元的赌注,全部输光, 游戏结束

看得师傅的
date: 2019-7-21
'''
from random import randint

money = 1000
game = False

while money > 0:
    print('money = %d' %money)
    game = False
    while True:
        debt = int(input('请下注:'))
        if debt > 0 and debt <= money:
            break
    first = randint(1, 6) + randint(1, 6)
    print('玩家摇出了%d点' % first)
    if first == 7 or first == 11:
        print('you won!')
        money += debt
    elif first == 2 or first == 3 or first == 12:
        print('you lost!')
        money -= debt
    else:
        game = True

    while game:
        then = randint(1, 6) + randint(1, 6)
        print('玩家摇出了%d点' % then)
        if then == 7:
            print('you lost')
            money -= debt
            game = False
        elif then == first:
            print('you won')
            money += debt
            game = True
            
print('哈哈,你破产了, 游戏结束!')
        
        
        
```
